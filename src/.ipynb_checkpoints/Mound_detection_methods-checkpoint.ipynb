Mound Detection Techniques & Image manipulation 
-----------------------------------
In this notebook I will be comparing the different methods for several blob techniques and with each one comparing several image manipulation to see what method serves us best. 

```python
%matplotlib inline 
from math import sqrt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh, canny 
from skimage.exposure import equalize_adapthist, equalize_hist
from skimage.color import rgb2gray
from skimage import exposure
import matplotlib.pyplot as plt
from collections import Counter
import operator
from PIL import Image
import numpy as np
import os
```

```python

plt.close('all')
images = ['0-3.png']
filepath = '../../Capstone_images/NN_ready_images'

for p in images:
    im = Image.open(filepath + '/' + p)
    im.load()
    im_data = np.asarray(im)
    grey_im = rgb2gray(im_data)

    #Laplacian of Gaussian (LoG) method
    blobs_log = blob_log(grey_im, min_sigma=1, max_sigma=10, threshold=.1, overlap=.1)
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

    # Determinant of Hessian method
    blobs_doh = blob_doh(grey_im, min_sigma=1, max_sigma=10, threshold=.002, overlap=0.1)
    blobs_doh[:, 2] = blobs_doh[:, 2] * sqrt(2)

    #Difference of Gaussian (DoG) method 
    blobs_dog = blob_dog(grey_im, threshold=.1, min_sigma = 1, max_sigma=10, overlap = 0.1)
    blobs_dog[:, 2] = blobs_dog[:,2] * sqrt(2) 

    blobs_lst = [blobs_log, blobs_doh, blobs_dog]
    colors = ['yellow', 'lime', 'red']
    titles = ['LoG Blobbing', 'DoH Blobbing', 'DoG Blobbing']

    order = zip(blobs_lst, colors, titles)
    f, ax = plt.subplots(1,4, figsize=(18, 16))
    
    

    for i, (blob, color, title) in enumerate(order):
        ax[i].set_title(title)
        ax[i].imshow(grey_im, interpolation='nearest')
        for b in blob:
            y, x, r = b
            c = plt.Circle((x,y), r, color=color, linewidth=2, fill=False)
            ax[i].add_patch(c)
        ax[i].set_axis_off()

    plt.tight_layout()
    plt.show()
```

Laplace of Gaussian
------------------
Uses bright on dark blobs only
Given an image, a Gaussian kernel(similarity function) with a small std deviation higher center value than surrounding values, then a low number of smoothing is done. Lager std deviations result in a larger kernel and bluring more of the photo, the values times the wieghts are averaged and given to the middle value. Then a Laplace(a function with respect to x & y, take the partial second derivative of each for the intensities  and sum them) filter is applied to the blurded image before it is applited on the image. which is the second derivative of the image. Where an imgae has constant intensity (gradient is zero) the resulting LoG will be zero but there it isnt constant the values will be positive for dark and negative for light. 

Determinant of Hessian
---------------------
Can use either type of blob 
fastest of all three methods
Hessian matrix packages all the second partial derivatives of a function


Difference of Gaussian
--------------------
Uses bright on dark blobs only
faster than LoG
What the DoG does is it uses a gaussian kernel again computes a smoothed image and does this for two different kernel sizes. It then produces an image that is the difference of the two images to produce higher gradients and more detail depending on the kernel size.

Log Correction:
    Replaces the pixel value with its log counterpart, which has the effect of enhancing lower pixel values

```python

plt.close('all')
images = ['0-3.png']
filepath = '../../Capstone_images/NN_ready_images'

for p in images:
    im = Image.open(filepath + '/' + p)
    im.load()
    im_data = np.asarray(im)
    grey_im = rgb2gray(im_data)
    logarithmic_corrected = exposure.adjust_log(grey_im, 1)

    #Laplacian of Gaussian (LoG) method
    blobs_log = blob_log(logarithmic_corrected, min_sigma=1, max_sigma=10, threshold=.1, overlap=.1)
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

    # Determinant of Hessian method
    blobs_doh = blob_doh(logarithmic_corrected, min_sigma=1, max_sigma=10, threshold=.002, overlap=0.1)
    blobs_doh[:, 2] = blobs_doh[:, 2] * sqrt(2)

    #Difference of Gaussian (DoG) method 
    blobs_dog = blob_dog(logarithmic_corrected, threshold=.1, min_sigma = 1, max_sigma=10, overlap = 0.1)
    blobs_dog[:, 2] = blobs_dog[:,2] * sqrt(2) 

    blobs_lst = [blobs_log, blobs_doh, blobs_dog]
    colors = ['yellow', 'lime', 'red']
    titles = ['LoG Blobbing', 'DoH Blobbing', 'DoG Blobbing']

    order = zip(blobs_lst, colors, titles)
    f, ax = plt.subplots(1,3, figsize=(18, 16))
    
    

    for i, (blob, color, title) in enumerate(order):
        ax[i].set_title(title)
        ax[i].imshow(logarithmic_corrected, interpolation='nearest')
        for b in blob:
            y, x, r = b
            c = plt.Circle((x,y), r, color=color, linewidth=2, fill=False)
            ax[i].add_patch(c)
        ax[i].set_axis_off()

    plt.tight_layout()
    plt.show()
```

Gamma Correction:
     Given some (V) a luminus value and raises it to the power of gamma, hence the name, times some constant. This is generally used to optimize useage of bits or bandwidth but is also used to display an image optimally by adjusting the brightness. The brightness eluminated by screen if precieved different by the human eye so before this gamma image file can be displayed it has to be corrected back where V^1/gamma 

```python


plt.close('all')
images = ['0-3.png']
filepath = '../../Capstone_images/NN_ready_images'

for p in images:
    im = Image.open(filepath + '/' + p)
    im.load()
    im_data = np.asarray(im)
    grey_im = rgb2gray(im_data)
    gamma_corrected = exposure.adjust_gamma(grey_im, 1)

    #Laplacian of Gaussian (LoG) method
    blobs_log = blob_log(gamma_corrected, min_sigma=1, max_sigma=10, threshold=.1, overlap=.1)
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

    # Determinant of Hessian method
    blobs_doh = blob_doh(gamma_corrected, min_sigma=1, max_sigma=10, threshold=.002, overlap=0.1)
    blobs_doh[:, 2] = blobs_doh[:, 2] * sqrt(2)

    #Difference of Gaussian (DoG) method 
    blobs_dog = blob_dog(gamma_corrected, threshold=.1, min_sigma = 1, max_sigma=10, overlap = 0.1)
    blobs_dog[:, 2] = blobs_dog[:,2] * sqrt(2) 

    blobs_lst = [blobs_log, blobs_doh, blobs_dog]
    colors = ['yellow', 'lime', 'red']
    titles = ['LoG Blobbing', 'DoH Blobbing', 'DoG Blobbing']

    order = zip(blobs_lst, colors, titles)
    f, ax = plt.subplots(1,3, figsize=(18, 16))
    
    

    for i, (blob, color, title) in enumerate(order):
        ax[i].set_title(title)
        ax[i].imshow(gamma_corrected, interpolation='nearest')
        for b in blob:
            y, x, r = b
            c = plt.Circle((x,y), r, color=color, linewidth=2, fill=False)
            ax[i].add_patch(c)
        ax[i].set_axis_off()

    plt.tight_layout()
    plt.show()
```

Histogram Equalization
--------------------

Contrast Stretching:
    Takes the min and max values in the image pixel range and streches the pixel values to encompass the the whole range, generally 0-255. 
    New_val = (pixel_value - lower) (hist_upper-hist_lower)/(p_95 - p_05) + lower

```python
#contrast streching
plt.close('all')
images = ['0-3.png']
filepath = '../../Capstone_images/NN_ready_images'

for p in images:
    im = Image.open(filepath + '/' + p)
    im.load()
    im_data = np.asarray(im)
    grey_im = rgb2gray(im_data)
     
    p2, p98 = np.percentile(grey_im, (10, 98))
    img_rescale = exposure.rescale_intensity(grey_im, in_range=(p2, p98))

    #Laplacian of Gaussian (LoG) method
    blobs_log = blob_log(img_rescale, min_sigma=1, max_sigma=10, threshold=.1, overlap=.1)
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

    # Determinant of Hessian method
    blobs_doh = blob_doh(img_rescale, min_sigma=1, max_sigma=10, threshold=.002, overlap=0.1)
    blobs_doh[:, 2] = blobs_doh[:, 2] * sqrt(2)

    #Difference of Gaussian (DoG) method 
    blobs_dog = blob_dog(img_rescale, threshold=.1, min_sigma = 1, max_sigma=10, overlap = 0.1)
    blobs_dog[:, 2] = blobs_dog[:,2] * sqrt(2) 

    blobs_lst = [blobs_log, blobs_doh, blobs_dog]
    colors = ['yellow', 'lime', 'red']
    titles = ['LoG Blobbing', 'DoH Blobbing', 'DoG Blobbing']

    order = zip(blobs_lst, colors, titles)
    f, ax = plt.subplots(1,3, figsize=(18, 16))
    
    

    for i, (blob, color, title) in enumerate(order):
        ax[i].set_title(title)
        ax[i].imshow(img_rescale, interpolation='nearest')
        for b in blob:
            y, x, r = b
            c = plt.Circle((x,y), r, color=color, linewidth=2, fill=False)
            ax[i].add_patch(c)
        ax[i].set_axis_off()

    plt.tight_layout()
    plt.show()
```

General Histogram Equalization:
    The histogram is composed by taking the number of pixel with insensity of (i) over the total number of pixels. The new pixel value is then calculated by taking the CDF of that value times its pixel intensity. This gives a larger range of pixel intensities. 

```python

plt.close('all')
images = ['0-3.png']
filepath = '../../Capstone_images/NN_ready_images'

for p in images:
    im = Image.open(filepath + '/' + p)
    im.load()
    im_data = np.asarray(im)
    grey_im = rgb2gray(im_data)
     
    
    img_eq = exposure.equalize_hist(grey_im)

    #Laplacian of Gaussian (LoG) method
    blobs_log = blob_log(img_eq, min_sigma=1, max_sigma=10, threshold=.1, overlap=.1)
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

    # Determinant of Hessian method
    blobs_doh = blob_doh(img_eq, min_sigma=1, max_sigma=10, threshold=.002, overlap=0.1)
    blobs_doh[:, 2] = blobs_doh[:, 2] * sqrt(2)

    #Difference of Gaussian (DoG) method 
    blobs_dog = blob_dog(img_eq, threshold=.1, min_sigma = 1, max_sigma=10, overlap = 0.1)
    blobs_dog[:, 2] = blobs_dog[:,2] * sqrt(2) 

    blobs_lst = [blobs_log, blobs_doh, blobs_dog]
    colors = ['yellow', 'lime', 'red']
    titles = ['LoG Blobbing', 'DoH Blobbing', 'DoG Blobbing']

    order = zip(blobs_lst, colors, titles)
    f, ax = plt.subplots(1,3, figsize=(18, 16))
    
    

    for i, (blob, color, title) in enumerate(order):
        ax[i].set_title(title)
        ax[i].imshow(img_eq, interpolation='nearest')
        for b in blob:
            y, x, r = b
            c = plt.Circle((x,y), r, color=color, linewidth=2, fill=False)
            ax[i].add_patch(c)
        ax[i].set_axis_off()

    plt.tight_layout()
    plt.show()
```

Adaptive Histogram Equalization:
    Similar to general HE but multiple histograms are composed for distinct regions of the image and used to compose the specific intensities for that region. Good for enhancing local contrast 

```python
#adaptive equalization
plt.close('all')
images = ['0-3.png']
filepath = '../../Capstone_images/NN_ready_images'

for p in images:
    im = Image.open(filepath + '/' + p)
    im.load()
    im_data = np.asarray(im)
    grey_im = rgb2gray(im_data)
     
    
    from skimage.morphology import closing, square
    
    img_adapteq = exposure.equalize_adapthist(grey_im, clip_limit=0.008)
    new_image = closing(img_adapteq, square(3))

    #Laplacian of Gaussian (LoG) method
    blobs_log = blob_log(new_image, min_sigma=1, max_sigma=10, threshold=.1, overlap=.1)
    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)

    # Determinant of Hessian method
    blobs_doh = blob_doh(new_image, min_sigma=1, max_sigma=10, threshold=.002, overlap=0.1)
    blobs_doh[:, 2] = blobs_doh[:, 2] * sqrt(2)

    #Difference of Gaussian (DoG) method 
    blobs_dog = blob_dog(new_image, threshold=.1, min_sigma = 1, max_sigma=10, overlap = 0.1)
    blobs_dog[:, 2] = blobs_dog[:,2] * sqrt(2) 

    blobs_lst = [blobs_log, blobs_doh, blobs_dog]
    colors = ['yellow', 'lime', 'red']
    titles = ['LoG Blobbing', 'DoH Blobbing', 'DoG Blobbing']

    order = zip(blobs_lst, colors, titles)
    f, ax = plt.subplots(1,3, figsize=(18, 16))
    
    

    for i, (blob, color, title) in enumerate(order):
        ax[i].set_title(title)
        ax[i].imshow(new_image, interpolation='nearest')
        for b in blob:
            y, x, r = b
            c = plt.Circle((x,y), r, color=color, linewidth=2, fill=False)
            ax[i].add_patch(c)
        ax[i].set_axis_off()

    plt.tight_layout()
    plt.show()
```
